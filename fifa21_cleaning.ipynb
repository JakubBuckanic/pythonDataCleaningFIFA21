```python
#Converiting whole column to cm
def convert_height(height):
    if 'cm' in height:
        return float(height.replace('cm', ''))
    elif '\'' in height:
        feet, inches = height.replace('"', '').split('\'')
        return float(feet) * 30.48 + float(inches) * 2.54
    else:
        return None

df['Height'] = df['Height'].apply(convert_height)
df[['Height']]
```

```python
#Checking results
df['Height'].unique()
```

```python
#Converiting whole column to kg
def convert_weight(weight):
    if 'kg' in weight:
        return float(weight.replace('kg', ''))
    elif 'lbs' in weight:
        return float(weight.replace('lbs', '')) * 0.453592
    else:
        return None

df['Weight'] = df['Weight'].apply(convert_weight)
df[['Weight']]
```

```python
#Checking results
df['Weight'].unique()
```

```python
#Stripping ★ symbol from 'W/F', 'SM', and 'IR' columns
df['W/F'] = df['W/F'].str.strip("★").astype(int)
df['SM'] = df['SM'].str.strip("★").astype(int)
df['IR'] = df['IR'].str.strip("★").astype(int)

#Renaming the columns with a star symbol
df = df.rename(columns={'W/F': 'W/F★', 'SM': 'SM★', 'IR': 'IR★'})
```

```python
#Checking results
df[['W/F★', 'SM★', 'IR★']]
```

```python
# Converting the column 'Joined' to date
df['Joined'] = pd.to_datetime(df['Joined'])
#Checking results
df[['Joined']]
```

```python
#Changing 'Wage', 'Value', 'Release Clause' to int type, removing sigh '€', replacing 'K' and 'M', with zeroes
columnsKM=['Wage', 'Value', 'Release Clause']

for c in columnsKM:
    df[c] = df[c].str.strip('€')
    d = {'M': '*1000000', 'K': '*1000'}
    s_convert = df[c].replace(d, regex=True).map(pd.eval)
    df[c] = s_convert.astype(int)
```
